```python
#reading in the raw crime dataset by removing the first column "ID"
crime = pd.read_csv("crime.csv", index_col=0)
crime.shape
```

```python
crime.dtypes
```

```python
crime.head(10)
```

```python
#Checking for missing data
crime.isnull().sum()
```

```python
#Dropping rows with missing values 
#Use Listwise Deletion

crime.dropna(axis=0, inplace=True)
crime.shape
```

```python
crime.isnull().sum()
```

```python
# Selecting the range of years I am intrested in which is Year 2020-2023

# Filtering crime dataset for rows with 'Year' between 2020 and 2023
crime_filtered = crime[(crime['Year'] >= 2020) & (crime['Year'] <= 2023)]

```

```python
#displaying the filtered years ranging from 2020-2024
crime_filtered.head(10)
```

```python
crime_filtered.shape
```

```python
# Create new DataFrame objects in order to seperate the date and the time in the date column.
crime_filtered_new = crime_filtered.copy()

# Extract date, month and time into separate columns
crime_filtered_new['Date_only'] = crime_filtered['Date'].dt.date
crime_filtered_new['Month_only'] = crime_filtered['Date'].dt.month
crime_filtered_new['Time_only'] = crime_filtered['Date'].dt.time
```

```python
#Viewing the new dataframe with the seperated columns. 
crime_filtered_new.head(10)
```

```python
# Convert numeric month to text using calendar module
month_mapping = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}

# Converting numeric month to text using dictionary mapping
crime_filtered_new['Month_only'] = crime_filtered_new['Month_only'].map(month_mapping)
```

```python
crime_filtered_new.head(10)
```

```python
# Droping the original 'Date' column
crime_filtered_new.drop(columns=['Date'], inplace=True)
crime_filtered_new.head(10)
```

```python
#Drop some Variables which is the "Updated on" column since I am not interested in the the date the record was updated. 
#I am only interested in the date and time the crime happened. 
crime_filtered_new= crime_filtered_new.drop(["Updated On"], axis=1)
crime_filtered_new.head(10)
```

```python
#Seperating the address on the Block column into block number and street name
# Split the Block column into Block Number and Street Name
crime_filtered_new[['Block Number', 'Street']] = crime_filtered_new['Block'].str.split(n=1, expand=True)

crime_filtered_new.head(10)

```

```python
#Dropping the original Block column since I have seperated it into block number and street
crime_filtered_new.drop(columns=['Block'], inplace=True)
crime_filtered_new.head(10)
```

```python
# exporting the clean dataset ranging from 2020-2023
crime_filtered_new.to_csv("chicago_crime_dataset.csv", index=False)
```

```python
#determining the primary type of crime in 2020

# Filtering the dataset to include only records from the year 2020
crimes_2020 = crime_filtered_new[crime_filtered_new['Year'] == 2020]

# Counting the occurrences of each primary type of crime and vewing the count with their frequencies. 

crime_counts = crimes_2020['Primary Type'].value_counts().reset_index()
crime_counts.columns = ['Primary Type', 'Frequency']

# Create an interactive bar plot using Plotly
fig = px.bar(crime_counts, x='Primary Type', y='Frequency', title='Primary Types of Crime in Chicago during Covid-19 Pandemic (2020)',
             labels={'Frequency': 'Frequency', 'Primary Type': 'Primary Type of Crime'},
             template='plotly_white')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
#Making static plot for the graph above

# Filtering the dataset to include only records from the year 2020
crimes_2020 = crime_filtered_new[crime_filtered_new['Year'] == 2020]

# Counting the occurrences of each primary type of crime
crime_counts = crimes_2020['Primary Type'].value_counts().reset_index()
crime_counts.columns = ['Primary Type', 'Frequency']

# Sort the dataframe by frequency
crime_counts = crime_counts.sort_values(by='Frequency', ascending=False)

# Create a static bar plot using matplotlib
plt.figure(figsize=(12, 8))
plt.bar(crime_counts['Primary Type'], crime_counts['Frequency'], color='skyblue')
plt.title('Primary Types of Crime in Chicago during Covid-19 Pandemic (2020)')
plt.xlabel('Primary Type of Crime')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Determining the top 5 crimes between 2020-2023

# Filtering the dataset to include records between 2020-2023
crimes_2020_to_2023 = crime_filtered_new[crime_filtered_new['Year'].between(2020, 2023)]

# Counting the occurrences of each primary type of crime from 2020-2023
crime_counts = crimes_2020_to_2023['Primary Type'].value_counts().reset_index()
crime_counts.columns = ['Primary Type', 'Frequency']

# Displaying the top 5 primary types of crime from 2020-2023
top_5_crimes = crime_counts.head(5)

# Create an interactive bar plot using Plotly
fig = px.bar(top_5_crimes, x='Primary Type', y='Frequency',
             title='Top 5 Primary Types of Crime in Chicago (2020-2023)',
             labels={'Frequency': 'Frequency', 'Primary Type': 'Primary Type of Crime'},
             template='plotly_white')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
# Group by 'Primary Type' and 'Year', and count the occurrences
crime_counts = crime_filtered_new.groupby(['Primary Type', 'Year']).size().unstack()

# Plot the frequency of different types of crimes over time
crime_counts.plot(kind='line', figsize=(12, 8), marker='o', linewidth=2)
plt.title('Frequency of Different Types of Crimes Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate the total number of crimes
total_crimes = len(crime_filtered_new)

# Calculate the number of crimes resulting in arrests
arrested_crimes = crime_filtered_new['Arrest'].sum()

# Calculate the proportion of crimes resulting in arrests
arrest_proportion = (arrested_crimes / total_crimes) * 100

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie([arrested_crimes, total_crimes - arrested_crimes], labels=['Arrested', 'Not Arrested'], autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Proportion of Crimes in Chicago Resulting in Arrests')
plt.show()
plt.savefig("proportion of crimes.png")
```

```python
# Filter arrests between 2020 and 2023
arrests_between_2020_2023 = crime_filtered_new[(crime_filtered_new['Year'] >= 2020) & (crime_filtered_new['Year'] <= 2023) & (crime_filtered_new['Arrest'] == True)]

# Grouping by year and counting the number of arrests per year
arrests_by_year = arrests_between_2020_2023.groupby(arrests_between_2020_2023['Year']).size().reset_index(name='Count')
arrests_by_year['Year'] = arrests_by_year['Year'].astype(str)

# Plotting interactive bar chart
fig = px.bar(arrests_by_year, x='Year', y='Count', labels={'Year': 'Year', 'Count': 'Number of Arrests'},
             title='Number of Arrests Made Between 2020-2023')
fig.update_layout(xaxis_tickangle=-45)
fig.show()

```
